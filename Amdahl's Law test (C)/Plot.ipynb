```python
from matplotlib import pyplot as plt
```

```python
wtime1 = 9.55092
wtime2 = 0.00286808
P = wtime1/(wtime1+wtime2)
lista = []

for threads in range(1,100001):
    f = 1/((1-P)+P/threads)
    lista.append(f)
    
len(lista)
```

```python
plt.plot(range(100000), lista)

plt.title('Lei de Amdahl')
plt.xlabel('Num Threads')
plt.ylabel('Speedup')

plt.show()
```
