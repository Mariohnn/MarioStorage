```python
from sklearn.model_selection import train_test_split

def embaralha(x,y):
    X_aux, X_emb, y_aux, y_emb = train_test_split(x, y, test_size=0.95 ,shuffle=True)
    r = (len(x)-len(x)*0.95)
    X_emb = X_emb.append(X_aux.iloc[0:int(r)])
    y_emb = y_emb.append(y_aux)
    return (X_emb, y_emb)
```

```python
modelo_SvmLin = svm.SVC(C=0.1, kernel='linear')
```

```python
def Leave1IndOut_Lin(Sample, Sick):
    y_pred = []
    y_total = Sick
    for i in range(40):
        X_test = Sample[i:i+1]
        y_test = Sick[i:i+1]
        x = pd.concat([Sample[0:i], Sample[i+1:40]])
        y = pd.concat([Sick[0:i], Sick[i+1:40]])
        X_train = x
        y_train = y
        X_train = sc.fit_transform(X_train)
        X_test = sc.fit_transform(X_test)
        modelo_SvmLin.fit(X_train, y_train)
        y_pred_aux = modelo_SvmLin.predict(X_test)
        y_pred = np.append(y_pred, y_pred_aux)
    return y_pred
    
def Majority_voting(matriz):
    mv = []
    for i in range(40):
        soma = 0
        for j in range(7):
            if matriz[j][i]==1:
                soma=soma+1
        if soma>3:
            mv.append(1)
        else:
            mv.append(0)
    return mv

def SVM_LIN(samples, sick):
    m = np.zeros((23,40))
    y_total = sick
    for i in range(23):
        m[i] = Leave1IndOut_Lin(samples[i], sick)
    y = Majority_voting(m)
    print("Acurácia:", accuracy_score(y_total, y))
        
```
