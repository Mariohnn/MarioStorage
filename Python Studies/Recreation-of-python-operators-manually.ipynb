```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics

%matplotlib inline

from tabulate import tabulate
```

```python
class iterator(type):
    
    def __iter__(self):
        return self.classiter()

    
class Coluna:
    
    def __init__(self, coluna, Nome):
        self.c = coluna
        self.name = Nome
        
    def __len__(self):
        return len(self.c)
    
    def __repr__(self):
        if type(self) == Coluna:
            dic = {self.name:self.c}
            return tabulate(dic, headers = 'keys')
    
    def show(self):
        return self.c
    
    def __getitem__(self, index):
        return self.c[index]
    
    def __setitem__(self, index, valor):
        self.c[index] = valor
    
    def __str__(self):
        return str(self.c)
    
    def __list__(self):
        return list(self.c)
    
    def max(self):
        return self.valor.max()

    def min(self):
        return self.valor.min()

    def mean(self):
        return statistics.mean(self.c)
    
    def __add__(self, coluna):
        "same as a + b"
        try:
            if(len(self.c)!=len(coluna.c)):
                raise TypeError("TypeError: O tamanho dos vetores deve ser o mesmo!")
            return Coluna(list([a + b for a, b in zip(self.c, coluna.c)]))
        except TypeError:
            print("TypeError: O tamanho dos vetores deve ser o mesmo!")
            raise
    
    def __lt__(self, coluna):
        if self.c < coluna.c:
            return True
        else:
            return False
    
    def __mul__(self, coluna, *args):
        'same as a * b'
        try: 
            if isinstance(coluna, int):
                return Coluna(list([a * coluna for a in self.show()]))
            else:
                if(len(self.c)!=len(coluna.show())):
                    raise TypeError("TypeError: O tamanho dos vetores deve ser o mesmo!")   
                else:
                    return Coluna(list([a * b for a, b in zip(self.show(), coluna.show())]))
        except TypeError:
            print("TypeError: O tamanho dos vetores deve ser o mesmo!")
            raise
    
    def __iter__(self, index):
        'same as a[b]'
        try:
            for idx,_ in enumerate(self.show()):
                print(self.show()[idx])
        except Exception:
            print("erro!")
            raise
    
    def __and__(self, coluna):
        return self.c & coluna.c

    def __or__(self, coluna):
        return self.c | coluna.c

    def __lt__(self, coluna):
        return self.c < coluna.c

    def __le__(self, coluna):
        return self.c <= coluna.c

    def __eq__(self, coluna):
        return self.c == coluna.c

    def __ne__(self, coluna):
        return self.c != coluna.c

    def __ge__(self, coluna):
        return self.c >= coluna.c

    def __gt__(self, coluna):
        return self.c > coluna.c

    def __sub__(self, coluna):
        "same as a - b"
        try:
            if(len(self.c)!=len(coluna.c)):
                raise TypeError("TypeError: O tamanho dos vetores deve ser o mesmo!")
            return Coluna(list([a - b for a, b in zip(self.c, coluna.c)]))
        except TypeError:
            print("TypeError: O tamanho dos vetores deve ser o mesmo!")
            raise
    
    def __truediv__(self, coluna):
        "same as a/b"
        try:
            if(len(self.c)!=len(coluna.c)):
                raise TypeError("TypeError: O tamanho dos vetores deve ser o mesmo!")
            return Coluna(list([a/b for a, b in zip(self.c, coluna.c)]))
        except TypeError:
            print("TypeError: O tamanho dos vetores deve ser o mesmo!")
            raise
    
    def __rmul__(self, coluna):       
        return self * coluna
    
    def cmul(self, num):
        try:
            return Coluna(list([a * num for a in self.show()]))
        except TypeError:
            print("TypeError: O tamanho dos vetores deve ser o mesmo!")
            raise
        
    def vmul(self, coluna):
        try:
            if(len(self.c)!=len(coluna.c)):
                raise TypeError("TypeError: O tamanho dos vetores deve ser o mesmo!")
            return Coluna(list([a * b for a, b in zip(self.show(), coluna.c)]))
        except TypeError:
            print("TypeError: O tamanho dos vetores deve ser o mesmo!")
            raise
            
    def plt(self):
        plt.plot(self.c, kind='bar', alpha=0.50, label='Temperature')
        plt.show()
```
